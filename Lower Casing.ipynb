```python
#This is a jupyter notebook to demonstrate a text preprocessing technique (Lower Casing)

# Converting text to lower casing

import pandas as pd #Pandas is the python library used to load and csv files into readable table
```

```python
data = pd.read_csv('sample.csv') #func read_csv resolves data arrangement to be presented table style
data.head()  #func head prints the first 5 rows of the table
```

```python
df = data[['text']] #text column is assigned to df
df['text'] = df['text'].astype(str) # all data in the column to be converted to strings
df.head()
```

```python
df['text_lower'] = df['text'].str.lower() #Here the data to be used is converted so all words will have same case formatting
df.head()
```
