```python
# using pandas library to load csv file 


import pandas as pd

df = pd.read_csv('sample.csv')
df.head()
```

```python
# spellchecker is python package used to sort spelling corrections

from spellchecker import SpellChecker
```

```python
spell = SpellChecker()


def corrections(text):
    """
    This function helps to correct misspelled words
    found in the text column
    """
    corrected_words = []
    bad_words = spell.unknown(text.split())
    for word in text.split():
        if word in bad_words:
            correction = spell.correction(word)
            corrected_words.append(correction if correction else word)
        else:
            corrected_words.append(word)
    return " ".join(corrected_words)

df['corrected_words'] = df['text'].apply(corrections)
df.head()
```
