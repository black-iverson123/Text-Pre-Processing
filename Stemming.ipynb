```python
# importing the pandas library to load csv file to table
import pandas as pd

df = pd.read_csv('sample.csv')
df.head()
```

```python
#will be employing the use of a stemming algorithm from the nltk package

from nltk.stem.porter import PorterStemmer
```

```python
stem = PorterStemmer()

# defining a function to implement the algorithm
def word_stem(text):
    """Applies Porter Stemming to each word in the input text."""
    stem = PorterStemmer()
    return " ".join([stem.stem(word) for word in text.split()])

df['stemmed_words'] = df['text'].apply(word_stem)
df.head()
```
